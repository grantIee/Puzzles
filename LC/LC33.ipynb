# 33. Search in Rotated Sorted Array:

*Suppose an array sorted in ascending order is rotated at some pivot unknown to you beforehand.*

*(i.e., `[0,1,2,4,5,6,7]` might become `[4,5,6,7,0,1,2]`).*

*You are given a target value to search. If found in the array return its index, otherwise return -1.*

*You may assume no duplicate exists in the array.*

*Your algorithm's runtime complexity must be in the order of O(log n).*

**Example 1:**
```python
Input: nums = [4,5,6,7,0,1,2], target = 0
Output: 4
```

**Example 2:**
```python
Input: nums = [4,5,6,7,0,1,2], target = 3
Output: -1
```

````python
# Code For Attempt 1:
class Solution:
    
    def binsearch(self, nums, target, right, left):
        """
        :type nums: List[int]
        :type target: int
        :type right: int
        :type left: int
        :type: int
        """
        index = right//2
        guess = nums[index]
        print(target - guess)
        if target - guess > 0:
            self.binsearch(nums, target, right, left // 2)
        elif target - guess < 0:
            self.binsearch(nums, target, right // 2, left)
        elif target - guess == 0:
            return index
        else:
            return -1

            
 
    def search(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        nums = sorted(nums)
        left = 0
        right = len(nums) - 1
        return self.binsearch(nums, target, right, left)
    
        
        
    
            
#Test
arr = [4,5,6,7,0,1,2]
target = 0

#Implement
test = Solution()
result = test.search(arr, target)
print(result)
            
            
        
        
        
````

````python
# Code For Attempt 2:
class Solution:
    def search(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        while True:
            try: 
                r = nums.index(target)
                break
            except ValueError:
                r = -1
                break

        return r
    
#Test 
arr = [4,5,6,7,0,1,2]
target = 0

#Implement
test = Solution()
result = test.search(arr, target)
print(result)
````
